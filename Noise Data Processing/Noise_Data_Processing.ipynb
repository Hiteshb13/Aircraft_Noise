```python
import pandas as pd
```

**Load Excel File for Noise Data**

```python
# Load Noise Monitoring Data
noise_data = pd.read_excel("Kopie von Fluglärmdaten 2022 (002).xlsx", sheet_name=["MP1","MP2","MP3","MP4","MP5","MP6","MP7","MP8","MP9"], header=3)  # Read all sheets
combined_noise_df = pd.concat(noise_data.values(), ignore_index=True)  # Merge all sheets

print("\n Noise Data Loaded:")
print(combined_noise_df.info())
print("\n Sample Noise Data:")
print(combined_noise_df.head())
```

```python
# Check Missing Values
print("Missing Values in Combined Noise Data:")
print(combined_noise_df.isnull().sum())
```

```python
display(combined_noise_df)
```

**Timestamp Handling**

```python
# Convert 'ATA/ATD' to datetime format
combined_noise_df["ATA/ATD"] = pd.to_datetime(combined_noise_df["ATA/ATD"], format='%d.%m.%Y %H:%M:%S')

# Handle ambiguous times using 'ambiguous' argument
combined_noise_df["ATA/ATD"] = combined_noise_df["ATA/ATD"].dt.tz_localize(
    "Europe/Berlin", ambiguous='NaT', nonexistent='NaT'
).dt.tz_convert("UTC")

# Drop rows where localization failed
combined_noise_df = combined_noise_df.dropna(subset=["ATA/ATD"])

print(combined_noise_df["ATA/ATD"].head())
```

**Saving the Processed  & Cleaned Noise Data**

```python
combined_noise_df.to_csv("Noise_df.csv", index=False)
```
