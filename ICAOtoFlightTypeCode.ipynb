```python
import joblib
import pandas as pd

# Load ADS-B Data (Trajectory Data)
adsb_data = joblib.load("data_2022_january.joblib")

df_adsb = adsb_data.data
```

```python
# Extract the date part (YYYY-MM-DD) for grouping
df_adsb["date"] = df_adsb["timestamp"].dt.date

# Sort by time (to maintain trajectory order)
df_adsb = df_adsb.sort_values(by=["icao24", "date", "timestamp"])

# Display processed data
display(df_adsb.head())
```

```python
# Create a unique flight identifier by grouping by (icao24, date)
df_adsb["flight_id"] = df_adsb.groupby(["icao24", "date"]).ngroup()
```

```python
# Sort by time (to maintain trajectory order)
df_adsb.groupby(["icao24","day"]).timestamp.min()
```

```python
from traffic.data import aircraft

# Step 1: Group your flight data
grouped_data = df_adsb.groupby("flight_id").agg({
    "icao24": "first",
    "callsign": "first",
    "date": "first",
    "timestamp": ["min", "max"]
}).reset_index()

# Step 2: Rename columns
grouped_data.columns = ["flight_id", "icao24", "callsign", "date", "first_seen", "last_seen"]

# Step 3: Build ICAO24 → typecode mapping efficiently
unique_icao24s = grouped_data["icao24"].dropna().unique()

def get_typecode_safe(icao):
    info = aircraft.get(icao)
    return getattr(info, "typecode", None) if info else None

icao24_to_typecode = {icao: get_typecode_safe(icao) for icao in unique_icao24s}

# Step 4: Add typecode column to grouped data
grouped_data["typecode"] = grouped_data["icao24"].map(icao24_to_typecode)

# ✅ Done
print("Grouped flight data with typecode:")
display(grouped_data.head())
```

```python
import requests

def lookup_opensky_aircraft(icao24):
    url = f"https://opensky-network.org/api/metadata/aircraft/icao/{icao24.lower()}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print("Aircraft Info:")
        for k, v in data.items():
            print(f"{k}: {v}")
    else:
        print(f"Aircraft not found or API limit reached. Status code: {response.status_code}")
```

```python
# Load Noise Monitoring Data (Excel)
noise_data = pd.read_excel("Kopie von Fluglärmdaten 2022 (002).xlsx", sheet_name=["MP1","MP2","MP3","MP4","MP5","MP6","MP7","MP8","MP9"], header=3)  # Read all sheets
combined_noise_df = pd.concat(noise_data.values(), ignore_index=True)  # Merge all sheets

print("\n📌 Noise Data Loaded:")
print(combined_noise_df.info())
print("\n🔍 Sample Noise Data:")
print(combined_noise_df.head())
```

```python
# Check Missing Values
print("Missing Values in Combined Noise Data:")
print(combined_noise_df.isnull().sum())
```

```python
display(combined_noise_df)
```

```python
# Convert 'ATA/ATD' to datetime format
combined_noise_df["ATA/ATD"] = pd.to_datetime(combined_noise_df["ATA/ATD"], format='%d.%m.%Y %H:%M:%S')

# Handle ambiguous times using 'ambiguous' argument
combined_noise_df["ATA/ATD"] = combined_noise_df["ATA/ATD"].dt.tz_localize(
    "Europe/Berlin", ambiguous='NaT', nonexistent='NaT'
).dt.tz_convert("UTC")

# Drop rows where localization failed
combined_noise_df = combined_noise_df.dropna(subset=["ATA/ATD"])

print(combined_noise_df["ATA/ATD"].head())
```

```python
import pandas as pd

# Convert 'ATA/ATD' column to datetime format
combined_noise_df['ATA/ATD'] = pd.to_datetime(combined_noise_df['ATA/ATD'], errors='coerce')

# Define time range
start_time = "2022-01-01 00:00:00"
end_time = "2022-01-31 23:59:59"

# Filter dataset based on the time range
df_filtered_noise = combined_noise_df[
    (combined_noise_df['ATA/ATD'] >= start_time) & (combined_noise_df['ATA/ATD'] <= end_time)
]

# Display filtered dataset
print("\n✅ Filtered Data:")
print(df_filtered_noise.info())

```

```python
display(df_filtered_noise)
```

```python
from datetime import timedelta

time_window = timedelta(minutes=5)
matches = []

# Make sure both timestamps are in UTC
df_adsb["timestamp"] = pd.to_datetime(df_adsb["timestamp"], utc=True)
df_filtered_noise["ATA/ATD"] = pd.to_datetime(df_filtered_noise["ATA/ATD"], utc=True)

# Match each noise event to nearby flights
for _, noise in df_filtered_noise.iterrows():
    event_time = noise["ATA/ATD"]
    noise_type = noise.get("AC Type", None)

    candidate_flights = df_adsb[
        (df_adsb["timestamp"] >= event_time - time_window) &
        (df_adsb["timestamp"] <= event_time + time_window)
    ]

    if noise_type:
        candidate_flights = candidate_flights[candidate_flights["typecode"] == noise_type]

    if not candidate_flights.empty:
        for _, flight in candidate_flights.iterrows():
            matches.append({
                "noise_time": event_time,
                "noise_level": noise.get("Max dB(A)", None),
                "typecode": flight.get("typecode", None),
                "altitude": flight.get("altitude", None),
                "hour": event_time.hour,
                "flight_id": flight.get("flight_id", None)
            })
```

```python
df_labeled = pd.DataFrame(matches)
df_labeled.dropna(subset=["noise_level", "altitude", "typecode"], inplace=True)

# Optional: encode time features
df_labeled["hour"] = df_labeled["hour"].astype(int)
df_labeled = pd.get_dummies(df_labeled, columns=["typecode"], drop_first=True)
```
